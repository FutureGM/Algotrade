# Financial Ratios

## Key Terms:
### AD = Annual Dividend
### AI = Average Inventory
### Am = Amortization
### BR = Bonds Ratio
### BVPS = Book Value Per Share
### CA = Current Assets
### CBI = Convertible Bond Interest
### CD = Current Depreciation
### CF = Cash Flow
### CL = Current Liabilities
### COGS = Cost of Goods Sold
### CR = Current Ratio
### CSO = Common Shares Outstanding
### CSRFC = Common Shares Resulting From Conversion
### CSE = Common Stockholder's Equity
### CSR = Common Stock Ratio
### CY = Current Yield
### DE = Debt Equity Ratio
### DPR = Dividend Payout Ratio
### EBIT = Earnings Before Interest and Taxes
### EPS = Earning per Share
### EPCS = Earnings per Common Share
### FDEPS = Fully Diluted Earnings per Share
### ICR = Interest Coverage Ratio
### IE = Interest Expense
### ITR = Inventory Turnover Ratio
### MoPR = Margin of Profit Ratio
### MP = Market Price
### MSP = Market Stock Price
### NAVPB = Net Asset Value Per Bond
### NI = Net Income
### NIAPD = Net Income After Preferred Dividend
### NP = Net Profit
### NPR = Net Profit Ratio
### NS = Net Sales
### NWC = Net Working Capital
### OCS = Outstanding Common Shares
### PD = Preferred Dividend
### PE = Price/Earning Ratio
### PIC = Paid-In Capital
### PRE = Preferred Equity
### PV = Par Value
### PVB = Par Value Bonds
### PVC = Par Value Common (Stocks)
### QR or ATR = Quick Ratio or Acid Test Ratio
### RoCE = Return on Common Equity
### RE = Retained Earnings
### Rev = Revenue
### SE = Stockholder's Equity
### TL = Total Liabilities
### TLC = Total Long-Term Capitalization
### TOCS = Total Outstanding Common Shares
### TSE = Total Shareholder Equity

```python
#1. Working Capital measures liquidity/solvency.  
#Ability to pay short-term obligations. 
# + Number= Ability to pay, - Number = inabiity to pay
NWC = 'CA - CL'
#2. Curren Ratio gaugies ability to pay current liabilites: 
#Score > 1 = Ability to Pay. < 1 Inability to pay
CR = "CA ÷ CL"
#3. Quick Ratio (Acid-test):
#Stricter measurment to pay-short term than Net working capital
QR = "(CA-Inventory)÷CL"
#4. Common Stock Ratio:
# Measures portion of total capitalization that's common stockholders' equity
# For creditors: high ratio is good; shows not highly leveraged (deep debt)
CSR = "(PVC + RE + PIC)÷TLC(SE+Bonds)"
#5. Bond ratio:
#Opposite stock ratio.  Measures portion of total capital that is long-term 
#debt/leverage
#Lower ratio = using less leverage(debt), operating more conservatively.  
#Good ratio =<30-40%
BR = "PVB + TLC"

#6. Earnings Per Common Share:
#Measures portion of earnings available to common stockholders
#After preferred stock has received its dividend
EPCS = "(NI - PD) ÷ CSO"
#7. Fully Diluted Earnings Per Share:
#Theoretical Numbers. Measures earnings available to common if all convertible
#securities were converted to common stock. Smaller number than EPS.
# Company's earnings are divided among a greater number of common shares
FDEPS = "(NI + CBI) ÷ (OCS + CSRFC)"
#8. Price/Earnings Ratio
#Compares stock's price to its share of earnings this year
#Indicates how fairly priced stock is compared to other comprable stocks
PE = "AD ÷ EPS"
#9. Dividend Payout Ratio:
#Measures "generosity" of board of directors. 
# Measures portion of earnings which board chooses to distribute to shares
#Remainder of earnings kepy by board in retained earnings
DPR = "AD ÷ EPS"
#10. Current Yield
#Measures the bnefit realized by purchasing the stock at current market value
CY = "AD ÷ MSP"
#11. Cash Flow
# Actual Cash generated by operations (as opposed to NI reportable to IRS)
# By adding back 'noncash' deductions to NI, it calculates actual cash figure
CF = "NI(or Loss) + CD + Am"
#12.Debt/Equity Ratio
DE = "TL ÷ SE"
#13.Inventory Turnover Ratio
#Measures how quickly inventory is sold.  
#Higher number = faster speed,m ore efficient management, lower inventory risk
ITR = "COGS ÷ AI"
#14. Margin-of-Profit Ratio or Profit Ratio
# Measures how much of each revenue dollar is net income
# Higher ratio = Good/Better
MoPR = "NI ÷ Rev"
#15. Net Profit Ratio
NPR = "NP ÷ NS"
#16. Net Asset Value Per Bond
# Calculates a company's net asset value, or book value, on a bond
NAVPB = ""
#17.Interest Coverage Ratio
#Measures how easily the entity can pay interest on its outstnading bonds.
#Higher number = Good/Better
ICR = "EBIT ÷ IE"
#18.Book value per share
# Liquidating value per common stock share. Book value left for common stock
BVPS = "(TSE - PRE) ÷ TOCS"
#19. Return on Common Equity
#Measures return which the corporation achieves on common stockholder's equity
# Higher return = better management
RoCE = "NIAPD ÷ CSE"
```
